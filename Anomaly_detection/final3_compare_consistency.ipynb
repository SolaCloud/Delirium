```python
import pandas as pd
ForestIsolation_sliding_02 = pd.read_csv('../output/Anomaly_delirium/forest_isolation_sliding_0.2contamination/forest_isolation_anomaly_data.csv')
ForestIsolation_sliding_01 = pd.read_csv('../output/Anomaly_delirium/forest_isolation_sliding_0.1contamination/forest_isolation_anomaly_data.csv')


df_LSTM_05sd = pd.read_csv('../output/Anomaly_delirium/LSTM_10days_baseline_0.5sd/LSTM_anomaly_data.csv')
df_LSTM_10sd = pd.read_csv('../output/Anomaly_delirium/LSTM_10days_baseline_1sd/LSTM_anomaly_data.csv')
df_LSTM_15sd = pd.read_csv('../output/Anomaly_delirium/LSTM_10days_baseline_1.5sd/LSTM_anomaly_data.csv')



print(ForestIsolation_sliding_02.shape)
print(ForestIsolation_sliding_01.shape)

print(df_LSTM_05sd.shape)
print(df_LSTM_10sd.shape)
print(df_LSTM_15sd.shape)
```

# 0.5sd in LSTM

```python
df_ForestIsolation = ForestIsolation_sliding_02.copy()
df_LSTM = df_LSTM_05sd.copy()

df_LSTM['date'] = pd.to_datetime(df_LSTM['date'])
df_LSTM['date'] = df_LSTM['date'].dt.strftime('%m-%d')


merged_df = pd.merge(
    df_ForestIsolation,
    df_LSTM[['patient_id', 'date']],  # only needed columns
    on=['patient_id', 'date'],                       # merge keys
    how='inner'                                       # or 'inner', 'right', 'outer'
)

rate_1  = merged_df.shape[0]/df_ForestIsolation.shape[0]
print(rate_1)
rate_2  = merged_df.shape[0]/df_LSTM.shape[0]
print(rate_2)

set_a = set(zip(df_ForestIsolation["patient_id"], df_ForestIsolation["date"]))
set_b = set(zip(df_LSTM["patient_id"], df_LSTM["date"]))

intersection = set_a & set_b
union = set_a | set_b
jaccard_index = len(intersection) / len(union)

print(jaccard_index)
```
