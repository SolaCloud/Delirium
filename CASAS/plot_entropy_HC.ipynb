```python
import pandas as pd
df_HC = pd.read_csv('casas_output/activity_entropy_rates_ALL_189.csv')
df_HC['Groups'] = 'health-control (N = '+str(len(df_HC.patient_id.unique()) )+')'
df_HC
```

```python
import os

# Path to the directory
folder_path = '../../../CASAS dataset/data/'

# Get a list of all CSV files in the directory
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Extract file names without the '.csv' extension
file_names_without_extension = [os.path.splitext(f)[0] for f in csv_files]
```

```python
df_TIHM = pd.read_csv('../output/data_cleaned/activity_entropy_rates.csv')
df_TIHM['Groups'] = 'patients (N = '+str(len(df_TIHM.patient_id.unique()) )+')'
df_TIHM
```

```python
# combine two dataframe
combined_df = pd.concat([df_HC, df_TIHM], ignore_index=True)
combined_df 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot for the two groups
sns.violinplot(x='Groups', y='entropy_rate', data=combined_df )

plt.savefig('casas_output/violin_plot_entropy_compare.png')
# Display the plot
plt.show()
```
