```python
import pandas as pd
df_raw = pd.read_csv('../output/data_cleaned/Mian_uti_m1_USED.csv')

df_raw
```

```python
df_uti = df_raw.copy()

df_uti['date'] = pd.to_datetime(df_raw['date'])
df_uti['dates'] = df_uti['date'].dt.date
df_uti
```

```python
# get daily count of agitation episodes
daily_counts = df_uti.groupby(['patient_id', 'dates']).size().reset_index(name='count')

daily_counts.columns = ['patient_id', 'date', 'uti_happen']
daily_counts.to_csv('../output/data_cleaned/uti_daily.csv', index=False)
daily_counts
```
