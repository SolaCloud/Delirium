```python
import pandas as pd
df_agitation_raw = pd.read_csv('../Dataset/Labels.csv')
df_agitation = df_agitation_raw.query(' type == "Agitation" ').copy()
df_agitation['date'] = pd.to_datetime(df_agitation['date'])
df_agitation['dates'] = df_agitation['date'].dt.date
df_agitation
```

```python
# get daily count of agitation episodes
daily_counts = df_agitation.groupby(['patient_id', 'dates']).size().reset_index(name='count')

daily_counts.columns = ['patient_id', 'date', 'agitation_counts']
daily_counts.to_csv('../output/data_cleaned/agitation_daily_counts.csv', index=False)
daily_counts
```

```python
# Group by patient_id and count unique time when agitation happened
new_df = df_agitation.groupby('patient_id')['date'].nunique().reset_index()
new_df.columns = ['patient_id', 'number_agitation']

new_df.to_csv('../output/data_cleaned/agitation_number.csv', index=False)
new_df
```
