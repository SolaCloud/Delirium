```python
import pandas as pd

bout_df = pd.read_csv('../output/data_cleaned/sleep_night_bouts_detailed.csv')
bout_df['start_time'] = pd.to_datetime(bout_df['start_time'])
bout_df['end_time'] = pd.to_datetime(bout_df['end_time'])

bout_df
```

```python
def assign_sleep_score(duration):
    if duration < 25:
        return 0  # restless sleep
    elif 25 <= duration < 45:
        return 1  # moderate sleep
    else:
        return 2  # good sleep

# Apply the function to create the 'sleep_score' column
bout_df['sleep_score'] = bout_df['total_duration'].apply(assign_sleep_score)
bout_df
```

```python
sleep_summary = bout_df.groupby(['patient_id', 'sleep_date'])['sleep_score'].agg(['sum', 'mean']).reset_index()
sleep_summary 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale the column
#### negated to change the direction
sleep_summary['scaled_sleep_quality_sum'] = -scaler.fit_transform(sleep_summary[['sum']])
sleep_summary['scaled_sleep_quality_mean'] = -scaler.fit_transform(sleep_summary[['mean']])
sleep_summary.to_csv('../output/sleep_score/sleep_quality_score_by_duration.csv', index=False)

sleep_summary
```
