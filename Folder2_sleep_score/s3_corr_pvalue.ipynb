```python
library(dplyr)
df1 = read.csv('../output/sleep_score/sleep_quality_score_by_duration.csv')
df1 = df1 %>% 
    dplyr::group_by(patient_id) %>% 
    dplyr::summarize(
        Sleep_quality_sum = mean(scaled_sleep_quality_sum, na.rm = TRUE),  # Calculate mean for each group
        Sleep_quality_mean = mean(scaled_sleep_quality_mean, na.rm = TRUE)     # Calculate sum for each group
  )
head(df1)
dim(df1)
```

```python
library(dplyr)
df2 = read.csv('../output/sleep_score/sleep_regularity_score.csv')
df2 = df2 %>% 
    dplyr::group_by(patient_id) %>% 
    dplyr::summarize(
        Regularity_cosine_similarity = mean(sleep_regularity_cs, na.rm = TRUE),  # Calculate mean for each group
        Irregularity_Euclidean_distance = mean(sleep_irregularity_dis, na.rm = TRUE)     # Calculate sum for each group
  )
head(df2)
dim(df2)
```

```python
agitation_number = read.csv('../output/data_cleaned/agitation_number.csv')
head(agitation_number)
dim(agitation_number)
```

```python
df_merge = merge(df1, df2, by = 'patient_id', all = TRUE)
df_merge = merge(agitation_number,df_merge, by = 'patient_id', all = TRUE)
df_merge = df_merge %>% filter(! ((is.na(Sleep_quality_sum) | is.na(Regularity_cosine_similarity))))
df_merge[is.na(df_merge)] <- 0
df_merge
dim(df_merge)
```

```python
cor(df_merge[,2:6], method = "kendall", use = 'complete.obs')
```

```python
cor.test(df_merge$number_agitation, df_merge$Sleep_quality_sum, method = "kendall", use = 'complete.obs')
```

```python
cor.test(df_merge$number_agitation, df_merge$Sleep_quality_mean, method = "kendall", use = 'complete.obs')
```

```python
cor.test(df_merge$number_agitation, df_merge$Regularity_cosine_similarity, method = "kendall", use = 'complete.obs')
```

```python
cor.test(df_merge$number_agitation, df_merge$Irregularity_Euclidean_distance, method = "kendall", use = 'complete.obs')
```
